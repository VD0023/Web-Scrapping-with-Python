Importing Libraries

```python
import pandas as pd
import numpy as np
import requests 
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

Creating empty lists

```python
Job_Name=[]
Job_Location=[]
Job_Link=[]
```

For Multiple Pages

```python
pages = np.arange(1,51,3)
```

```python
for page in pages:
  page = requests.get("https://careers.blackboard.com/search/jobs/in/country/india?page="+str(page)+"")
  soup = BeautifulSoup(page.text,'html.parser')
  EBook_data = soup.findAll('div', attrs = {'class': 'jobs-section__item page-section-small'})
  sleep(randint(2,8))
  for store in EBook_data:
    name = store.h5.a.text
    Job_Name.append(name)
    location = store.find('div',class_ ='large-4 columns').text.replace('\n','').replace(' ','').replace('Location:','')
    Job_Location.append(location)
    div_link=store.find("div",class_="large-8 columns")
    anchor_link=div_link.find("a")
    Job_Link.append(anchor_link.get("href"))



   # link = store.h5.a.get('herf')
   # Job_Link.append("https:///careers.blackboard.com" + str(link))
```

Creating a df

```python
Book_List = pd.DataFrame({ "Job": Job_Name, "Job Location": Job_Location, "Job Link": Job_Link})
```
